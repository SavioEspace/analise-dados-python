### (3) Tipos de gráficos adequados

- **Barra (ranking):** para comparar `canal`, `uf`, `categoria`, `segmento` por `receita` e `lucro`.
- **Linha (série temporal):** para evolução de `receita` e `lucro` ao longo da coluna `data` (ex.: por mês).
- **Dispersão:** relação entre `receita` x `lucro` (ou `custo` x `lucro`) para identificar padrões/outliers.
